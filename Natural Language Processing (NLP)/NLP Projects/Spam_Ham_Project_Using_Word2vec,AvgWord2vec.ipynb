```python
import pandas as pd
messages=pd.read_csv('../SpamClassifier-master/smsspamcollection/SMSSpamCollection',sep='\t',names=["label","message"])
messages.head()
```

```python
from nltk.stem import WordNetLemmatizer
import re
```

```python
import nltk
nltk.download('wordnet')
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(len(messages)):
  review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
  review=review.lower()
  review=review.split()
  review=[lemmatizer.lemmatize(word) for word in review]
  review=' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
nltk.download('punkt_tab')
```

```python
words=[]
for sentence in corpus:
  sent_token=sent_tokenize(sentence)
  for word in sent_token:
    words.append(simple_preprocess(word))
```

```python
words
```

```python
import gensim
```

```python
from gensim.models import Word2Vec,KeyedVectors
```

```python
# Lets train word2vec from scratch
model=Word2Vec(words)
```

```python
model.wv.index_to_key
```

```python
model.corpus_count
```

```python
model.wv.similar_by_word('good')
```

```python
model.wv['good'].shape
```

```python
words[0]
```

```python
import numpy as np
```

```python
def avg_word2vec(doc):
    embeddings = [model.wv[word] for word in doc if word in model.wv.index_to_key]
    if len(embeddings) == 0:
        return np.zeros(model.vector_size, dtype=np.float32)
    return np.mean(embeddings, axis=0, dtype=np.float32)
```

```python
!pip install tqdm
```

```python
from tqdm import tqdm
```

```python
X=[]
for i in tqdm(range(len(words))):
    X.append(avg_word2vec(words[i]))
```

```python
X
```

```python
len(X)
```

```python
# Independent Features
X_new=np.array(X)
```

```python
X_new.shape
```

```python
X_new[0].shape
```

```python
# Dependent Feature
y = messages[list(map(lambda x: len(x)>0 ,corpus))]
y=pd.get_dummies(y['label'])
y=y.iloc[:,0].values
```

```python
y.shape
```

```python
# Final Independent Feature
df=pd.DataFrame()
for i in range(0,len(X)):
  df=pd.concat([df,pd.DataFrame(X[i].reshape(1,-1))],ignore_index=True)
```

```python
df.shape
```

```python
df.head()
```

```python
# Independent Features
X=df
```

```python
X.head()
```

```python
# Dependent Feature
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

```python
X_train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(f"Accuracy Score: {accuracy_score(y_test,y_pred)}")
print(f"Classification Report: {classification_report(y_test,y_pred)}")
```

### Prediction of new Email

```python
message1="""Hey Amir! Hope you are doing well. This email is to inform you that you have selected for ML Engineer job at Turing"""
message2="""Hey Amir! Congrats You have won a lottery of 5000 Million PKR. Send us 1000 Rs. as token so that we can transfer your prize amount. and we will kill you"""
```

```python

message1_vector=avg_word2vec(message1)
message2_vector=avg_word2vec(message2)
```

```python
output=classifier.predict(message1_vector.reshape(1,-1))
if(output[0]==True):
    print("Ham")
else:
    print("Spam")
```

```python
output=classifier.predict(message2_vector.reshape(1,-1))
if(output[0]==True):
    print("Ham")
else:
    print("Spam")
```
