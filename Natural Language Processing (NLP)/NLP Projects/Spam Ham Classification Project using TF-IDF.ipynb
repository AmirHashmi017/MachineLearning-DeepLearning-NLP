## Spam Ham Classification Project using TF-IDF

```python
import pandas as pd
messages=pd.read_csv('../SpamClassifier-master/smsspamcollection/SMSSpamCollection',
                     sep='\t',names=['label','message'])
messages
```

```python
import re
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review if word not in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# Ouput Feature
y=pd.get_dummies(messages['label'])
```

```python
y
```

```python
# Dependent Feature
y=y.iloc[:,0].values
```

```python
y
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
# Indedpendent Feature
X_train=vectorizer.fit_transform(X_train).toarray()
X_test=vectorizer.transform(X_test).toarray()
```

```python
X_train
```

```python
vectorizer.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model=MultinomialNB()
spam_detect_model.fit(X_train,y_train)
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(f"Accuracy Score: {accuracy_score(y_test,y_pred)}")
print(f"Classification Report: {classification_report(y_test,y_pred)}")
```

### Prediction of new Email

```python
message1="""Hey Amir! Hope you are doing well. This email is to inform you that you have selected for ML Engineer job at Turing"""
message2="""Hey Amir! Congrats You have won a lottery of 5000 Million PKR. Send us 1000 Rs. as token so that we can transfer your prize amount. and we will kill you"""
```

```python
# Convert to Vector
vectorized_message1=vectorizer.transform([message1]).toarray()
vectorized_message2=vectorizer.transform([message2]).toarray()
```
