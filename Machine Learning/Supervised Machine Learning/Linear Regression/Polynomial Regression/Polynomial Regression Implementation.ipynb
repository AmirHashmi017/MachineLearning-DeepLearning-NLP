## Polynomial Regression Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating no-linear dataset
X=6* np.random.rand(100,1)-3
Y=0.5*X**2+1.5*X+2+np.random.rand(100,1)
# quadratic equation used- y=0.5x^2+1.5x+2+outliers
plt.scatter(X,Y,c='g')
plt.xlabel("X")
plt.ylabel("Y")
```

```python
# Step 01: Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# Implmenting Linear Regression on it
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(X_train,Y_train)
```

```python
# Performing Prediction
Y_pred=regression.predict(X_test)
```

```python
# Checking Model Accuracy or performance by R2 score
from sklearn.metrics import r2_score
r2score=r2_score(Y_test,Y_pred)
print("R Sqaure: ",r2score)
```

#### You can see that the model with linear regression is giving less accuracy

```python
# Let's visualize this model
plt.plot(X_train,regression.predict(X_train),color='r')
plt.scatter(X_train,Y_train)
plt.xlabel('X')
plt.ylabel('Y')
```

#### Now we will solve this problem using Polynomial Regression and increase the accuracy of our model

```python
# Let's apply polynomail transformation
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2,include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
X_train_poly
# So it converted it into 2 Degree Polynomial
```

```python
# Now we will again apply Linear Regression on it
from sklearn.linear_model import LinearRegression
regression_p=LinearRegression()
regression_p.fit(X_train_poly,Y_train)
```

```python
# Predict the output
Y_pred=regression_p.predict(X_test_poly)
```

```python
# Checking Model Accuracy or performance by R2 score
from sklearn.metrics import r2_score
r2score=r2_score(Y_test,Y_pred)
print("R Sqaure for Polynomial Regression: ",r2score)
```

#### You can see that how the accuracy of model is improved

```python
# I have three Coeffiecients
print(regression_p.coef_)
```

```python
# Now let's visualize this model
plt.scatter(X_train,regression_p.predict(X_train_poly))
plt.scatter(X_train,Y_train)
plt.xlabel("X")
plt.ylabel("Y")
```

```python
# Now if we make the polynomianl dgree 3
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3,include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.fit_transform(X_test)
```

## Pipeline Concepts

#### We use these Piplines to make function by which we can try different degrees effciently.In it we combine two models

#### Basically Pipline is used to combine Estimators

```python
from sklearn.pipeline import Pipeline 
```

```python
def poly_regression(degree):
    X_new=np.linspace(-3,3,200).reshape(200,1)
    poly=PolynomialFeatures(degree=degree,include_bias=True)
    lin_reg=LinearRegression()
    poly_regression=Pipeline([("poly_features",poly),
                             ("lin_reg",lin_reg)])
    poly_regression.fit(X_train,Y_train) ## ploynomial and fit of linear reression
    Y_pred_new=poly_regression.predict(X_new)
    Y_pred=poly_regression.predict(X_test)
    #plotting prediction line
    plt.plot(X_new, Y_pred_new,'r', label="Degree " + str(degree), linewidth=2)
    plt.plot(X_train, Y_train, "b.", linewidth=3)
    plt.plot(X_test, Y_test, "g.", linewidth=3)
    plt.legend(loc="upper left")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.axis([-4,4, 0, 10])
    plt.show()
    # Caluclating Accuracy 
    r2score=r2_score(Y_test,Y_pred)
    print("R Square: ",r2score)
```

```python
poly_regression(2)
```
