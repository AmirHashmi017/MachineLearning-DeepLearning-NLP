## Support Vector Classification (SVC) Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Create Dataset
from sklearn.datasets import make_classification
X,Y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0,random_state=40)
```

```python
X
```

```python
Y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=Y)
```

```python
# Traintest split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
svc=SVC(kernel="linear")
svc.fit(X_train,Y_train)
```

```python
svc.coef_,svc.intercept_
```

```python
# Prediction
Y_pred=svc.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(Y_test,Y_pred))
print(confusion_matrix(Y_test,Y_pred))
```

### RBF Kernel

```python
# For now you can see the accuracy is good but e ca improve it by trying different kernels
svc_kernel=SVC(kernel="rbf")
svc_kernel.fit(X_train,Y_train)
```

```python
# Prediction
Y_pred1=svc_kernel.predict(X_test)
Y_pred1
```

```python
print(classification_report(Y_test,Y_pred1))
print(confusion_matrix(Y_test,Y_pred1))
```

Now you can see accuracy improved using rbf Kernel

### Polynomial Kernel

```python
svc_poly=SVC(kernel="poly")
svc_poly.fit(X_train,Y_train)
Y_pred2=svc_poly.predict(X_test)
print(classification_report(Y_test,Y_pred2))
print(confusion_matrix(Y_test,Y_pred2))
```

It is not giving that good accuracy

### Sigmoid Kernel

```python
svc_sigmoid=SVC(kernel="sigmoid")
svc_sigmoid.fit(X_train,Y_train)
Y_pred3=svc_sigmoid.predict(X_test)
print(classification_report(Y_test,Y_pred3))
print(confusion_matrix(Y_test,Y_pred3))
```

```python
svc_sigmoid.intercept_
```

Again accuracy is not that good so we will use RBF Kernel

## HyperParameter Tuning

```python
from sklearn.model_selection import GridSearchCV
svc=SVC()
param_grid={'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001],'kernel':['rbf']}
grid=GridSearchCV(estimator=svc,param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,Y_train)
```

```python
grid.best_params_
```

```python
# Prediction
Y_pred4=grid.predict(X_test)
print(classification_report(Y_test,Y_pred4))
print(confusion_matrix(Y_test,Y_pred4))
```
